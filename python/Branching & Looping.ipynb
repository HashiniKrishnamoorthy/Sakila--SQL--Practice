# LOOPING
## We have FOR and While, both iterates through the DS and give instructions
#### FOR loop is used when we know the condition to iterrates and where to stop 
#### WHILE loop is used when we know only the condition, when condition which is not breakable then it be infinite loop

```python
n = int(input())
sum=0
for i in range(1,n):
    sum=sum+i
    i=i+1
print(sum)
```

```python
n=[1,2,3,4,5]
l=[]
for i in range(len(n)-1,-1,-1):
    l.append(n[i])
print(l)
```

```python
n=int(input())
while n!=0:
    print(n,end=" ")
    n=n-1
```

```python
n=int(input())
while n!=0:
    t=n%10
    print(t,end="")
    n=n//10
```

# Break and Continue 
#### Break is used inside a loop to stop the loop immediately, even if it has not finished all iterations.
#### Continue is used inside a loop to skip the current iteration and move to the next one.

```python
# Loop 1 to 10,Stop the loop when the number becomes 5.
n= int(input())
t=int(input())
for i in range(1,n):
    print(i,end =" ")
    if i==t:
        break
```

```python
#Write a loop that prints each number until it finds given number, then stops
l=[2,4,5,7,9,10,13,16,17]
t=int(input())
for i in range(0,len(l)):
    if l[i] == t:
        print(f"it is at the postion of {i}")
        break
else:
    print("not found")
```

```python
#Write a loop that prints numbers from 1 to n, but skips all even numbers.

n= int(input())
for i in range(1,n+1):
    if i%2==0:
        continue
    print(i,end=" ")
```

```python
# print the numbers from 1 to 20 and skip the number 17 and not print multiple of 3
for i in range(1,20):
    if i == 17:
        break
    if i % 3 == 0:
        continue
    print(i,end =" ")
```
