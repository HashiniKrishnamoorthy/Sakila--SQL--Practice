# TUPLE
#### Immutable, unchanagable and allows duplicates because it is indexed

```python
t = (1,'data',30.54,'tuple')
print(t)
print(type(t))
```

```python
print(t[0])
print(t[3])
print(t[0:4])
print(t[-1])
print(t[-4:-1])
```

```python
t[0]='change'
```

```python
# if we want to change the tuple elements we need to convert to list
# then after changing same the list methods append,insert and extend can be done
temp= list(t)
print(type(temp))
temp[1]='change'
t=tuple(temp)
print(type(t))
print(t)
```

```python
# to remove we can follow the same process
temp= list(t)
print(type(temp))
temp.pop(0)
t=tuple(temp)
print(type(t))
print(t)
```

# SETS
#### immutable, don not allow the duplicates ,unchangeable
#### Set items do not have a definit order, set items can appear in a different order every time and cannot use them you referred by index or key.
#### we cannot change the item of set after creation of it nas once set is created, we cannot change them but we can add new items and remove Items

```python
s={1,'data',2,3,4}
print(s)
print(type(s))
```

```python
# if we want to access the elements in the set we cannot do it by index but we can loop through them

for i in s:
    if i == 'data':
        print(i)
```

```python
# adding elements to the sets can be done by two methods
# with add() and update()
sets = {'apple','kiwi','orange'}
sets.add('banana')
print(sets)
```

```python
# if we want copy elements from another sets it can be done by update()
# this can be used to join lists, tuples or dictonaries to the set

l=[1,2,5,6,7,8,'list']
t=(34,56,75,'tuple')
s={1,'data',2,3,4}
sets.update(s)
sets.update(l)
sets.update(t)
print(sets)
```

# LIST AND DICTIONARY COMPREHENSION 

squares = []
for x in range(5):
    squares.append(x*x)

print(squares)

```python
squares = [x*x for x in range(5)]
print(squares)
```

```python
square_dict = {}
for x in range(5):
    square_dict[x] = x*x

print(square_dict)
```

```python
square_dict = {x: x*x for x in range(5)}
print(square_dict)
```

### Shallow copy and Deep copy

```python
import copy
l=[1,2,3,4]
l2= copy.deepcopy(l)
l2[2]= 24
print(l2)
print(l)
```

```python
import copy

a = [1, 2, [3, 4]]
b = copy.copy(a)   # shallow copy

b[2][0] = 99   # change inside nested list

print("a:", a)
print("b:", b)
```

```python
import copy

a = [1, 2, [3, 4]]
b = copy.deepcopy(a)   # deep copy

b[2][0] = 99   # change inside nested list

print("a:", a)
print("b:", b)
```
