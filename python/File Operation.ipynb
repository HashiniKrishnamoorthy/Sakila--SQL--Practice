# FILE OPERATION
### Text Files

```python
with open ("fileoperation.txt","w") as file:
    file.write("this is file operation.\n")
    file.write("this include all operation.\n")

    
```

```python
with open("fileoperation.txt","r") as file:
    create = file.read()
    print("file content:")
    print(create)
```

```python
with open ("fileoperation.txt","a") as file:
    file.write("this operation adds the words.\n")
    file.write("we can see the line are not overwritten.\n")
```

```python
# this operation will see the line by line operations in the txt file
with open("fileoperation.txt","r") as file:
     for line in file:
         print("line:", line.strip())
    
```

```python
# getting first characters, includes spaces and special characters
with open("fileoperation.txt","r") as file:
    print("first 9 characters:" ,file.read(9))
```

```python
#read line by line without for loop
# this will print line by line , per each line we need to call it , for 1000 line with wont work
# if we dont have any line it wont print , I have shown it down
with open("fileoperation.txt","r") as file:
    print("line1: ", file.readline().strip())
    print("line2: ", file.readline().strip())
    print("line3: ", file.readline().strip())
    print("line4: ", file.readline().strip())
    print("line5: ", file.readline().strip())
```

```python
# to print content in form of list we can use this 
with open("fileoperation.txt","r") as file:
    line = file.readlines()
    print("file content in list:")
    print(line)
```

```python
# copying the files

with open("fileoperation.txt","r") as file , open("copying.txt", "w") as cpy:
          for line in file:
              cpy.write(line)
```

```python
with open("copying.txt", "r") as cpy:
    content = cpy.read()
    print(content)
```

```python
# read + write mode 
# the seek() operation goes to the position we mention , it will overwrites the content 
with open ("fileoperation.txt","r+") as f:
    f.seek(20)
    f.write(" this read and write overtieen ")
```
