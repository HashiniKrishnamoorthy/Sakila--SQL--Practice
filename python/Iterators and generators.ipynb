# GENERATORS

It is special type of iterators
it uses YIELD keyword instead of return , 
YIELD pauses and stores in the memory and again wherever it is called it again resumes starts where it is stopped
why YIELD instead of RETURN -> the RETURN ends the function , wherever the data is needed again it must me start from first 
                               where as the YIELD it pause the function results and stores the results ,whereever it is needed it resumes from it to                                   next 
ex: fabbanoci series

```python
 
def gensqur(n):
    for i in range(n):
        yield pow(i,3)
```

```python
gensqur(12) # here the generator is intiallied and stores the result at particular memory location
```

```python
for x in gensqur(9):
    print(x)
```

## MAP() AND REDUCE() 

the map() which takes multiple input and gives to a function to generate multiple output 
reduce() which take multiple input and gives to a function and generates a single output

1,2,3,4 ----> map() to x*x ----> generates 1,4,9,16 
1,2,3,4 -----> reduce() to sum() ----> 10

```python
s = ['cat','dog','pig']

r = list(map(lambda x:x[0] ,s))
print(r)
```

### LAMBDA function
here lambda is function , instead loops we can use them , which reduces the looping to a single line
which is used when we want it dynamically
the syntax is
lambda( arguments : expression)
more examples

```python
a=[1,2,3,4]
b=(4,5,6,7)
x=lambda a,b: list(a) + list(b)
print(x(a,b))
```

```python
even_odd = lambda x: "Even" if x % 2 == 0 else "Odd"
print(even_odd(8)) 
```

```python
# reduce() function
# to use reduce() function we need to import it from function tools
from functools import reduce
l=[2,5,7,0,-2]
max_f = lambda a,b: a if a>b else b
print(reduce(max_f,l))
```

```python

from functools import reduce

max_s = lambda a,b: a if pow(a,2) > pow(b,2) else b

l= [5,2,7,3,1]

print(reduce(max_s,l))
```

### FILTER()
To use filter() function , the return value must be true (boolean function)

```python
nums = [5, 12, 7, 20, 3]

result = list(filter(lambda x: x > 10, nums))
print(result)
```

```python
l=[3,5,7,8,24,56]

result = list(filter(lambda x: True if x%2 ==0 else False, l))
print(result)
```
