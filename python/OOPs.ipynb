#### Abstraction

NOW the main part in OOPS (TYPES)
1.Abstraction:
 Showing only essential things to the user
 Hiding how things are done internally

```python
from abc import ABC, abstractmethod

class PaniPuriShop(ABC):
    def __init__(self):
        self.shop_name = "Tasty Pani Puri"   # Public
        self._price = 30                    # Protected

    @abstractmethod
    def serve(self, plates):
        pass
```

```python
class StreetPaniPuri(PaniPuriShop):
    def serve(self, plates):
        print(f"Serving {plates} plates of pani puri at {self.shop_name}")
        print(f"Price per plate: {self._price}")
```

```python
shop = StreetPaniPuri()
shop.serve(2)
print(shop._price)
```

```python
class PaniPuri:
    def __init__(self):
        self.shop_name = "Tasty Pani Puri"   # Public variable
        self._price = 30                    # Protected variable
        self.__recipe = "Secret Masala"     # Private variable 

    def get_price(self):
        return self._price

    def serve(self):
        print("Serving delicious pani puri ")
```

```python
class PaniPuriShop:
    def __init__(self):
        # Composition: Shop HAS-A PaniPuri
        # removing the useless details from concept
        self.item = PaniPuri()

    def serve_customer(self):
        print("Customer arrives at:", self.item.shop_name)
        self.item.serve()
        print("Price:", self.item.get_price())
```

```python
shop = PaniPuriShop()
shop.serve_customer()
```

```python
def get_discount(self, percent):
    discount = self._price * (percent / 100)
    final_price = self._price - discount
    return final_price
```

```python
PaniPuri.get_discount = get_discount
```

####  Polymorphism and DuckTyping

```python
class Dog:
    def sound(self):
        return "Bark"

class Cat:
    def sound(self):
        return "Meow"

# Polymorphism
def animal_sound(animal):
    print(animal.sound())

dog = Dog()
cat = Cat()

animal_sound(dog)  
animal_sound(cat)  
```

```python
# for duck typing
# if different class has same methods, the python allows to called as same ( qucks is same as duck)
class Duck:
    def quack(self):
        print("Quack!")

class Person:
    def quack(self):
        print("I can imitate a duck!")

def make_it_quack(obj):
    obj.quack()   

d = Duck()
p = Person()

make_it_quack(d)  
make_it_quack(p) 
```

### Instance , class andd static methods

```python
class Student:
    school_name = "ABC School"  # Class variable

    def __init__(self, name, age):
        self.name = name    
        self.age = age      

    # Instance method (needs self)
    def show_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

    # Class method (needs cls)
    @classmethod
    def show_school(cls):
        print(f"School: {cls.school_name}")

    # Static method (no self or cls)
    @staticmethod
    def greet():
        print("Welcome to the school!")

# Using the class
s1 = Student("John", 16)
s2 = Student("Alice", 15)

# Instance method
s1.show_info()  
s2.show_info()  
# Class method
Student.show_school()  
s1.show_school()       

# Static method
Student.greet()  
s2.greet()       
```

#### Decorators

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator # using the function name as @name_syntax
def say_hello():
    print("Hello!")

say_hello()
```

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int

    def __post_init__(self): # it is a special methoid in data class
        if self.age < 0:
            raise ValueError("Age cannot be negative")
        self.is_adult = self.age >= 18 

s = Student("John", 20)
print(s.is_adult)  # True/false 
```
