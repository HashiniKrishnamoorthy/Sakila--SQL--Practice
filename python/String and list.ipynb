# LISTS
##### They are Muttable, ordered and allows duplicates 
##### They can have have differeent king od datatypes in the list 

```python
l =[1,'data',2.4, 'string']
print(l)
```

```python
print(l[1])
print(l[0])
print(l[-2]) # negative indexing in the list 
```

```python
# Sciling in the List
print(l[1:3])
print(l[1:5:1])
print(l[-3:-1:1])
print(l[:-1])
```

```python
# checking the data in the list
if 2 in l:
    print('yes')
else:
    print('no')
```

```python
#adding the element in the list
# we can use Append, Insert and extend functions

l.insert(4,'insert') #insert do it by index number
print(l)
l.append('append') #adds the element to last and as insert also
print(l)

l2=['extend'] # it helps to copy the one list into the existing list 
l.extend(l2)
print(l)
```

```python
# Removing Elements from list 
# remove,pop,del,clear functions 

l.remove('data')
print(l)
```

```python
l.pop()
```

```python
print(l)
```

```python
del l[1]
print(l)
```

```python
 l.clear() # it emty the list
print(l)
```

```python
del l # it completely deletes the list
print(l)
```
