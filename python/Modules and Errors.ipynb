# Error Handling

```python
x = int(input("Enter a number: "))
result = 10 / x
print(result)
```

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Please enter a valid number")
finally:
    print("Done")
```
