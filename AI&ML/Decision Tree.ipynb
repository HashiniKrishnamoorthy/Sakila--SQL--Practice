## Decision Tree

A decision tree is a supervised learning algorithm used for classification and regression that models decisions in the form of a tree structure, where data is repeatedly split using if–else conditions on features, and the final outcome is predicted at the leaf nodes based on the learned rules.



- A **Decision Tree** is a supervised machine learning algorithm used for both **classification** and **regression**.
- It is one of the most **versatile and powerful** algorithms, especially for complex datasets.
- Decision trees are **easy to understand, interpret, and visualize**.
- When combined with **ensemble techniques** (such as Random Forest and Boosting), their performance improves significantly.
- The algorithm works by **dividing the dataset into a tree-like structure** using a series of **rules and conditions**.
- Each **internal node** represents a decision based on a feature.
- Each **branch** represents the outcome of a decision.
- Each **leaf node** represents the final prediction.
- The tree begins with a **root node**, selected using a splitting criterion (e.g., Gini impurity or Information Gain).
- Nodes are split into **child nodes** based on decision rules.
- If a node satisfies the condition, it becomes a **leaf node** and no further splitting occurs.
- If a node does not satisfy the condition, it is **split further** using another rule.
- The splitting process continues **recursively**.
- Tree growth stops when:
  - all conditions are satisfied, or  
  - a predefined stopping criterion (such as maximum depth) is reached.
- The **depth of the tree** controls model complexity and helps prevent overfitting.

### Decision Tree for Regression 

Decision Tree Regression is a supervised learning algorithm that predicts a continuous output by recursively dividing the feature space into distinct and non-overlapping regions. For any observation falling into a region, the predicted value is the mean of the target values of the training observations within that region. The splits are chosen using a greedy recursive binary splitting approach to minimize the Residual Sum of Squares (RSS).

##  Steps in Decision Tree Regression

1. Start with the entire dataset at the root node.  
2. Select a feature and a split value to divide the data into two regions.  
3. Compute the Residual Sum of Squares (RSS) for the split.  
4. Choose the split that gives the minimum RSS.  
5. Split the node into two child nodes.  
6. Repeat the splitting process recursively for each child node.  
7. Stop splitting when a stopping criterion is reached.  
8. Predict the output using the mean value of the target variable in each leaf node.

### Post-Pruning (Backward Pruning)
Post-pruning is a method where:
- The decision tree is first grown **completely**
- Then **non-important branches are removed**

A validation or cross-validation dataset is used to check:
- whether removing a branch improves model performance

If pruning improves or does not reduce accuracy, the branch is removed; otherwise, it is kept.



### Pre-Pruning (Forward Pruning)
Pre-pruning stops the growth of the decision tree **early**.
It prevents unnecessary branches from being created by applying conditions such as:
- maximum depth
- minimum samples per node
- minimum impurity decrease

This reduces overfitting by stopping splits that are not useful.


## Classification Trees

Decision trees used for **categorical (qualitative) target variables** are called **classification trees**.  
Unlike regression trees (which predict numerical values), classification trees predict **class labels**.

- **Regression trees** use **Residual Sum of Squares (RSS)** to split nodes.
- **Classification trees** use:
  - Classification Error Rate
  - **Entropy**
  - **Gini Impurity**

The main goal is to create **pure nodes**, where most observations belong to the same class.

## Entropy

**Entropy** is a measure of **randomness or impurity** in a dataset.

- High entropy → data is mixed (more randomness)
- Low entropy → data is clean (mostly one class)

The objective while splitting nodes is to **reduce entropy**.

 **Entropy formula** =  **- (P*log2(p))**
  
  where p is probability 

  ## Information Gain

Information Gain measures the **reduction in entropy** after a dataset is split.

Formula:
**Gain(T,X) = Entropy(T)before - Entropy(T,X)after**

where T is parent node,x is spliliting node from T 


## Gini Impurity

Gini impurity measures how often a randomly chosen observation would be **incorrectly classified**.
Gini Impurity checks how often a randomly selected sample would be mislabeled if assigned by class probability. It is computationally simple and used in tree-based classifiers.

It reflects how **mixed** the classes are in a node.

### Steps to Select Root Node in Decision Tree (Classification)

### Using Gini Impurity

1. Start with the complete dataset at the root node.
2. Identify all possible features that can be used for splitting.
3. For each feature:
   - Split the dataset based on its unique values.
   - Calculate the Gini impurity for each resulting group.
4. Compute the **weighted Gini impurity** for each feature.
5. Compare the weighted Gini values of all features.
6. Select the feature with the **lowest weighted Gini impurity** as the root node.

<img src ="Maths steps -1(classifier).png" width ="500" >
### Using Entropy and Information Gain

1. Start with the complete dataset at the root node.
2. Calculate the **entropy of the parent node**.
3. For each feature:
   - Split the dataset based on its unique values.
   - Calculate the entropy for each resulting group.
4. Compute the **weighted entropy** for each feature.
5. Calculate **Information Gain** for each feature.
6. Select the feature with the **highest Information Gain** as the root node.

## Different Algorithms for Decision Tree

###  ID3 (Iterative Dichotomiser 3)
- ID3 is a decision tree algorithm used for classification.
- It uses **Information Gain** as the criterion for selecting the root node and splits.
- It works **only with categorical attributes**.
- It does not support continuous features directly.

### C4.5
- C4.5 is an improved version of ID3
- It can handle both **continuous and discrete attributes**.

### CART (Classification and Regression Trees)
- CART is the **most popular decision tree algorithm**.
- It can be used for both **classification and regression** problems.
- It uses **Gini Impurity** as the default criterion for classification.
- Entropy can also be used instead of Gini impurity.
- CART is widely used in practical implementations, including **Python (scikit-learn)**.

### Gini Impurity vs Entropy
- Both Gini impurity and Entropy measure node impurity.
- Using either of them usually gives similar results.
- Gini impurity is faster and easier to compute because it does not involve logarithms.
- Hence, **CART uses Gini impurity by default**.

## Advantages of Decision Tree

- Can be used for both classification and regression problems.
- Very easy to understand and interpret.
- Decision rules are clearly visible.
- Complex trees become easy to understand when visualized.
- No need for feature scaling or normalization.

## Disadvantages of Decision Tree

- Small changes in data can cause large changes in the tree structure.
- High chance of overfitting, especially for deep trees.
- Training can be slower compared to some other algorithms.


## Cross-Validation – Simple Definitions (Class Notes)

### Cross-Validation
Cross-validation is a technique used to evaluate a machine learning model by training it on one part of the data and testing it on unseen data, helping to detect and reduce overfitting.


### Hold-Out Method
The hold-out method splits the dataset into two parts: a training set and a test set.  
The model is trained on the training set and evaluated on the test set.

### k-Fold Cross-Validation
In k-fold cross-validation, the dataset is divided into k equal parts.  
The model is trained k times, each time using one part as the test set and the remaining k−1 parts as the training set.  
The final performance is the average of all k results.
<img src ="K-fold Error.png" width ="500" >


### Leave-One-Out Cross-Validation (LOOCV)
LOOCV is a special case of k-fold cross-validation where k equals the total number of observations.  
Each observation is used once as test data while the remaining observations are used for training.


### Bias–Variance Tradeoff in Cross-Validation
- Hold-out method has high variance because results depend on a single split.
- LOOCV has low bias but high variance.
- k-fold cross-validation provides a balance between bias and variance.
- k-fold CV (usually k = 5 or 10) is most commonly used in practice.



```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Class":  ["I","I","II","II","III","III","III","I","II","III"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Female","Male","Female","Male"],
    "Hostel": ["Yes","No","Yes","No","Yes","Yes","No","Yes","No","Yes"]
})
```

```python
X = df[["Class", "Gender"]]
y = df["Hostel"]

print("Dataset:")
display(df)
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), ["Class", "Gender"])
    ],
    remainder="drop"
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# ginni impurity calculation
tree_gini = DecisionTreeClassifier(criterion="gini", random_state=42)

model_gini = Pipeline(steps=[
    ("prep", preprocess),
    ("tree", tree_gini)
])

model_gini.fit(X_train, y_train)
pred_gini = model_gini.predict(X_test)

print("\n--- Baseline Decision Tree (Gini) ---")
print("Accuracy:", accuracy_score(y_test, pred_gini))
print("Confusion matrix:\n", confusion_matrix(y_test, pred_gini))
print("classification report:\n",classification_report(y_test, pred_gini, zero_division=0))
```

```python
#Train Decision Tree (Entropy) - compare criterion
tree_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)

model_entropy = Pipeline(steps=[
    ("prep", preprocess),
    ("tree", tree_entropy)
])

model_entropy.fit(X_train, y_train)
pred_entropy = model_entropy.predict(X_test)

print("\n--- Decision Tree (Entropy) ---")
print("Accuracy:", accuracy_score(y_test, pred_entropy))
print("Confusion matrix:\n", confusion_matrix(y_test, pred_entropy))
print("Classification report:\n", classification_report(y_test, pred_entropy, zero_division=0))
```

```python
#cross-Validation (k-fold) to estimate generalization

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

cv_scores_gini = cross_val_score(model_gini, X, y, cv=cv, scoring="accuracy")
cv_scores_entropy = cross_val_score(model_entropy, X, y, cv=cv, scoring="accuracy")

print("\n Fold Cross-Validation:")
print("Gini CV accuracy scores:", np.round(cv_scores_gini, 3), "Mean:", np.round(cv_scores_gini.mean(), 3))
print("Entropy CV accuracy scores:", np.round(cv_scores_entropy, 3), "Mean:", np.round(cv_scores_entropy.mean(), 3))
```

```python
#Pre-pruning (Forward pruning) - stop tree early
#    Use constraints like max_depth, min_samples_leaf

tree_prepruned = DecisionTreeClassifier(
    criterion="gini",
    max_depth=2,           # limit depth
    min_samples_leaf=2,    # require at least 2 samples per leaf
    random_state=42
)

model_prepruned = Pipeline(steps=[
    ("prep", preprocess),
    ("tree", tree_prepruned)
])

pre_scores = cross_val_score(model_prepruned, X, y, cv=cv, scoring="accuracy")

print("\n Pre-pruning (max_depth=2, min_samples_leaf=2):")
print("CV accuracy scores:", np.round(pre_scores, 3), "Mean:", np.round(pre_scores.mean(), 3))
```

```python
# Post-pruning (Backward pruning) using cost-complexity pruning
#    Steps:
#  Fit a full tree
#  Get ccp_alpha path
#  CV over alphas, pick best alpha

# Fit preprocessing once to get numeric matrix for pruning path
X_encoded = preprocess.fit_transform(X)
# Full tree (no constraints)
full_tree = DecisionTreeClassifier(criterion="gini", random_state=42)
full_tree.fit(X_encoded, y)

path = full_tree.cost_complexity_pruning_path(X_encoded, y)
ccp_alphas = path.ccp_alphas
# Remove duplicate/very tiny values to keep it neat
ccp_alphas = np.unique(ccp_alphas)

alpha_scores = []
for a in ccp_alphas:
    clf = DecisionTreeClassifier(criterion="gini", random_state=42, ccp_alpha=a)
    # CV directly on encoded data here
    scores = cross_val_score(clf, X_encoded, y, cv=cv, scoring="accuracy")
    alpha_scores.append(scores.mean())

best_idx = int(np.argmax(alpha_scores))
best_alpha = float(ccp_alphas[best_idx])

print("\n Post-pruning (Cost-Complexity Pruning):")
print("Number of candidate alphas:", len(ccp_alphas))
print("Best ccp_alpha:", best_alpha)
print("Best CV mean accuracy:", round(alpha_scores[best_idx], 3))

# Train final pruned model as a pipeline (so it works on raw X)
tree_postpruned = DecisionTreeClassifier(criterion="gini", random_state=42, ccp_alpha=best_alpha)
model_postpruned = Pipeline(steps=[
    ("prep", preprocess),
    ("tree", tree_postpruned)
])
post_scores = cross_val_score(model_postpruned, X, y, cv=cv, scoring="accuracy")
print("Post-pruned CV accuracy scores:", np.round(post_scores, 3), "Mean:", np.round(post_scores.mean(), 3))


```

```python
# Visualize one tree (example: pre-pruned or post-pruned)
# Fit on full dataset for visualization (pre-pruned)
model_prepruned.fit(X, y)

# Get feature names after OneHotEncoding
ohe = model_prepruned.named_steps["prep"].named_transformers_["cat"]
feature_names = ohe.get_feature_names_out(["Class", "Gender"])

plt.figure(figsize=(12, 6))
plot_tree(
    model_prepruned.named_steps["tree"],
    feature_names=feature_names,
    class_names=model_prepruned.named_steps["tree"].classes_,
    filled=True,
    rounded=True
)
plt.title("Decision Tree (Pre-pruned) Visualization")
plt.show()
```
